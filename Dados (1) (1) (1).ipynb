```python
import datetime
df2['year'] = df2['date'].dt.year
# month
df2['month'] = df2['date'].dt.month
# day
df2['day'] = df2['date'].dt.day
# week of year
df2['week_of_year'] = df2['date'].dt.isocalendar().week

df2['year_week'] = df2['date'].dt.strftime( '%Y-%W' )

df2['competition_since'] = df2.apply( lambda x: datetime.datetime(year=x['competition_open_since_year'],month=x['competition_open_since_month'],day=1 ), axis=1 )
df2['competition_time_month'] = ( ( df2['date'] - df2['competition_since'] )/30).apply( lambda x: x.days ).astype( int )
df2['promo_since'] = df2['promo2_since_year'].astype( str ) + '-' +df2['promo2_since_week'].astype( str )
df2['promo_since'] = df2['promo_since'].apply( lambda x: datetime.datetime.strptime( x + '-1', '%Y-%W-%w' ) - datetime.timedelta( days=7 ) )
df2['promo_time_week'] = ( ( df2['date'] - df2['promo_since'] )/7 ).apply(lambda x: x.days ).astype( int )
# assortment
df2['assortment'] = df2['assortment'].apply( lambda x: 'basic' if x == 'a' else 'extra' if x == 'b' else 'extended' )
# state holiday
df2['state_holiday'] = df2['state_holiday'].apply( lambda x: 'public_holiday' if x == 'a' else 'easter_holiday' if x == 'b' else 'christmas' if x == 'c' else 'regular_day' )
```

```python
from scipy import stats as ss
```

```python
def cramer_v(x, y):
    cm = pd.crosstab(x, y).values
    n = cm.sum()
    r, k = cm.shape
    chi2 = ss.chi2_contingency(cm)[0]
    chi2corr = max(0, chi2 - (k - 1) * (r - 1) / (n - 1))
    kcorr = k - (k - 1) ** 2 / (n - 1)
    rcorr = r - (r - 1) ** 2 / (n - 1)
    return np.sqrt((chi2corr / n) / (min(kcorr - 1, rcorr - 1)))

# only categorical data
a = df4.select_dtypes( include='object' )
# Calculate cramer V
a1 = cramer_v( a['state_holiday'], a['state_holiday'] )
a2 = cramer_v( a['state_holiday'], a['store_type'] )
a3 = cramer_v( a['state_holiday'], a['assortment'] )
a4 = cramer_v( a['store_type'], a['state_holiday'] )
a5 = cramer_v( a['store_type'], a['store_type'] )
a6 = cramer_v( a['store_type'], a['assortment'] )
27
a7 = cramer_v( a['assortment'], a['state_holiday'] )
a8 = cramer_v( a['assortment'], a['store_type'] )
a9 = cramer_v( a['assortment'], a['assortment'] )
# Final dataset
d = pd.DataFrame( {'state_holiday': [a1, a2, a3],
'store_type': [a4, a5, a6],
'assortment': [a7, a8, a9] })
d = d.set_index( d.columns )
sns.heatmap( d, annot=True )
```

```python
import torch # é uma bilbioteca usada para criar rede neurais
import torch.nn as nn #usado para treinar e construir as camadas de redes neurais, como entrada, oculta e saída
import torch.optim as optim #algoritmos de otimização
from sklearn.model_selection import train_test_split #dividir entre treino e teste
from sklearn.preprocessing import StandardScaler #para transformar vetores de recursos brutos em uma representação que seja mais adequada para os estimadores e normalizar os dados
from torch.utils.data import DataLoader, TensorDataset, random_split

df = df6[cols_selected_boruta_full]
X = df6[cols_selected_boruta].values # converte os valores em um array numpy que contem apenas os valores das colunas
y = df6['sales'].values # faz o mesmo com esse

num_samples = 1000 #defini com 1000 amostras
num_features = len(cols_selected_boruta) # as features sao a qnt de valores que tem nas colunas do boruta
                    #linhas     #colunas
                    #1000      15( a quantidade das colunas dos dados la)
X = np.random.rand(num_samples, num_features).astype(np.float32)# aqui vamos gerar uma matriz de valores aleatorios que estão no intervalo de 0-1, sendo o tipo 'np.float32' por ser mais eficiente do que o 'np.64'
y = np.random.rand(num_samples, 1).astype(np.float32) #vamos gerar outra matriz aleatoria com apenas uma coluna e sendo as linhas as amostras aleatorias


X_tensor = torch.tensor(X)# usamos para criar variaveis que armazenam tensores
y_tensor = torch.tensor(y)


dataset = TensorDataset(X_tensor, y_tensor) #a classe 'TensorDataset' é usada para combinar os tensores e deixá-los mais praticos
train_size = int(0.8 * len(dataset)) #80% para treino e 20% para teste
val_size = len(dataset) - train_size # 20% para o teste
train_dataset, val_dataset = random_split(dataset, [train_size, val_size]) #dividimos o dataset em treino e teste do tamanho especificado

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True) # criar um iterador que percorre os dados em lotes, tendo o conjunto de treinamento em 32 pedaços, e o shufle embaralha para evitar ficar preso em padrões
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)# aqui nao embaralhamos pois estamos preocupados com o desempenho do modelo

# Definindo a Rede Neural
class SimpleNN(nn.Module):
    def __init__(self, input_dim, output_dim): #definindo as camadas da rede neural, recebendo como entrada posteriormente, o numero de features e o output_dim como saida
        super(SimpleNN, self).__init__()# chama a classe 'pai' , sendo uma classe 'filho'( subclasse) que deseja herdar as caracteristicas da classe dominante
        self.layer1 = nn.Linear(input_dim, 64) #CRIA A PRIMEIRA CAMADA LINEAR DA REDE NEURAL ( transformar os dados da entrada em dimensoes diferentes), produzindo 64 saídas
        self.relu = nn.ReLU() #cria uma camada de ativação , transforma os valores negativos em zero e mantém os valores positivos inalterados
        self.output = nn.Linear(64, output_dim) #cria a camada de saida da rede neural



    def forward(self, x): #definindo o forward da rede neural / passagem direta
        x = self.layer1(x) # passa os dados pela primeira camada
        x = self.relu(x)# aplica a função de ativação para essa camada, que foi armazenada em x
        x = self.output(x)  # passa pela saída
        return x

input_dim = num_features
output_dim = 1

model = SimpleNN(input_dim, output_dim)

# Definindo o otimizador e a função de perda
optimizer = optim.Adam(model.parameters(), lr=0.001) #vamos otimizar com o algoritmo Adam, com uma taxa de aprendizado de 0.001
criterion = nn.MSELoss() #definimos a função de perda com o MSELoss( Loss Quadrática Média (Mean Squared Error - MSE)),  dá um maior peso aos maiores erros, verifica a acuracia

# Treinamento
num_epochs = 50 #iterações completas

for epoch in range(num_epochs): #booleana de epoca em epoca
    model.train()  #inicia o modelo
    train_loss = 0.0 #inicia com perda 0
      # para cada iteração, ele vai ter um numero de características
    for inputs, targets in train_loader:
      # dividimos em lotes e usamos o paralelismo
        optimizer.zero_grad() #zera os gradientes acumulados
        outputs = model(inputs) #gera previsões do modelo a partir da passagem direta dos inputs(dados de entrada)
        loss = criterion(outputs, targets) #calculamos a função da perda entre os outputs( previsoes ) e as características de cada input(rotulos verdadeiros)
        loss.backward()# calcula os gradientes de função da perda em relação aos parametros do modelo (RETROPROPAGATION)ESTUDAAAAAr MAAAIS
        optimizer.step()#BACKPROPAGATION ESTUDAAAAR
        train_loss += loss.item() * inputs.size(0)#perda de treinameno acumulada para cada epoca, multiplica pelo tamanho do lot( input) para verificar a relação com os pesos

    train_loss /= len(train_loader.dataset)

#No entanto, após algumas épocas, parece estabilizar ou começar a aumentar ligeiramente. Isso pode indicar que o modelo está começando a overfit, ou seja, está se ajustando muito bem aos dados de treinamento, mas não está generalizando tão bem para dados não vistos.


    #avaliação do modelo
    model.eval() # Define o modelo no modo de avaliação (não treinamento)
    val_loss = 0.0  # Inicializa a perda de validação
    with torch.no_grad(): # Desliga o cálculo de gradientes
        for inputs, targets in val_loader:
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            val_loss += loss.item() * inputs.size(0)  # Acumula a perda de validação

    val_loss /= len(val_loader.dataset)

    print(f'Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss:.4f}, Validation Loss: {val_loss:.4f}')

# Avaliação final
model.eval()
with torch.no_grad():
    val_loss = 0.0
    for inputs, targets in val_loader:
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        val_loss += loss.item() * inputs.size(0)

val_loss /= len(val_loader.dataset)
print(f'Final Validation Loss: {val_loss:.4f}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def cross_validation( x_training, kfold, model_name, model, verbose=False ):
    mae_list = []
    mape_list = []
    rmse_list = []
    for k in reversed( range( 1, kfold+1 ) ):
        if verbose:
            print( '\nKFold Number: {}'.format( k ) )
        validation_start_date = x_training['date'].max() - datetime.timedelta(days=k*6*7)
        validation_end_date = x_training['date'].max() - datetime.timedelta(days=(k-1)*6*7)
# filtering dataset
        training = x_training[x_training['date'] < validation_start_date]
        validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]
# training and validation dataset
# training
        xtraining = training.drop( ['date', 'sales'], axis=1 )
        ytraining = training['sales']
# validation
        xvalidation = validation.drop( ['date', 'sales'], axis=1 )
        yvalidation = validation['sales']
# model
        m = model.fit( xtraining, ytraining )
# prediction
        yhat = m.predict( xvalidation )
# performance
        m_result = ml_error( model_name, np.expm1( yvalidation ), np.expm1(yhat ) )
# store performance of each kfold iteration
        mae_list.append( m_result['MAE'] )
        mape_list.append( m_result['MAPE'] )
        rmse_list.append( m_result['RMSE'] )
    return pd.DataFrame( {'Model Name': model_name,
'MAE CV': np.round( np.mean( mae_list ), 2 ).astype(str ) + ' +/- ' + np.round( np.std( mae_list ), 2 ).astype( str ),
'MAPE CV': np.round( np.mean( mape_list ), 2 ).astype( str ) + ' +/- ' + np.round( np.std( mape_list ), 2 ).astype( str ),
'RMSE CV': np.round( np.mean( rmse_list ), 2 ).astype( str ) + ' +/- ' + np.round( np.std( rmse_list ), 2 ).astype( str )}, index=[0] )

    # start and end date for validation
def mean_percentage_error( y, yhat ):
    return np.mean( ( y - yhat ) / y )
def mean_absolute_percentage_error( y, yhat ):
    return np.mean( np.abs( ( y - yhat ) / y ) )
def ml_error( model_name, y, yhat ):
    mae = mean_absolute_error( y, yhat )
    mape = mean_absolute_percentage_error( y, yhat )
    rmse = np.sqrt( mean_squared_error( y, yhat ) )
    return pd.DataFrame( { 'Model Name': model_name,
'MAE': mae,
'MAPE': mape,
'RMSE': rmse }, index=[0] )

```
